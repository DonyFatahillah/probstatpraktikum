```python
import pandas as pd

data_dony = pd.read_csv("praktikum1.csv")

data_dony
```

```python
subsetdata = data_dony[['Nama', 'Tinggi Badan']]

subsetdata
```

```python
data_dony.loc[data_dony['Tinggi Badan'] > 160, 'Tinggi Badan'] = 'Tinggi'
data_dony.loc[data_dony['Tinggi Badan'] != 'Tinggi', 'Tinggi Badan'] = 'Pendek'

print(data_dony)
```

```python
data_dony = data_dony.assign(Jurusan="Infor20")
data_dony = data_dony.assign(Fakultas="FTI")

print(data_dony)
```

```python
data_dony = data_dony.drop(columns=["Fakultas"])

print(data_dony)
```

```python
data_dony['Nama_Gender'] = data_dony['Nama'] + "_" + data_dony['Gender']

data_dony['Angkatan_TinggiBadan'] = data_dony['Angkatan'].astype(str) + "_" + data_dony["Tinggi Badan"].astype(str)

data_dony["Gabungan"] = data_dony["Nama_Gender"] + "_" + data_dony["Angkatan_TinggiBadan"]

print(data_dony)
```

```python
gabungan_1 = data_dony.iloc[0:5].sum()

gabungan_2 = data_dony.iloc[24:30].sum()

gabungan_1 = gabungan_1.astype(str)
gabungan_2 = gabungan_1.astype(str)

gabungan_akhir = gabungan_1 + gabungan_2

print(gabungan_akhir)
```
